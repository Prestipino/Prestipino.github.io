### Exercise 1: Moles to Grams Converter
`mass = molar_mass × moles`

### Exercise 13: Radioactive Decay
`m = m₀ × 0.5^(t / t₁/₂)`

```python
# SOLUTION: Exercise 13
initial_mass = float(input("Initial mass (g): "))
half_life_years = float(input("Half-life (years): "))
years = float(input("Time elapsed (years): "))

half_lives = years / half_life_years
remaining_mass = initial_mass * (0.5 ** half_lives)

print(f"After {years} years: {remaining_mass:.4f} g remaining")
```

### Exercise 15: Projectile Motion
Range: `R = (v² × sin(2θ)) / g`
Max height: `H = (v² × sin²θ) / (2g)`

```python
# SOLUTION: Exercise 15
import math

velocity = float(input("Initial velocity (m/s): "))
angle_degrees = float(input("Launch angle (degrees): "))
g = 9.81

# Convert to radians
angle_radians = math.radians(angle_degrees)

# Calculate range
range_distance = (velocity ** 2 * math.sin(2 * angle_radians)) / g

# Calculate max height
max_height = (velocity ** 2 * math.sin(angle_radians) ** 2) / (2 * g)

print(f"Range: {range_distance:.1f} meters")
print(f"Maximum height: {max_height:.1f} meters")
```

## Mini Project: Stoichiometry Calculator

**Reaction:** 2H₂ + O₂ → 2H₂O

```python
# SOLUTION: Mini Project
print("--- STOICHIOMETRY CALCULATOR ---")
print("Reaction: 2H₂ + O₂ → 2H₂O")

molar_mass_h2 = 2.016   # g/mol
molar_mass_h2o = 18.015 # g/mol

mass_h2 = float(input("Mass of H₂ (grams): "))

# Step 1: Convert mass H₂ to moles
moles_h2 = mass_h2 / molar_mass_h2

# Step 2: Stoichiometric ratio (1:1 from 2:2)
moles_h2o = moles_h2

# Step 3: Convert moles H₂O to grams
mass_h2o = moles_h2o * molar_mass_h2o

# Step 4: Print result
print(f"\nFrom {mass_h2} g of H₂:")
print(f"You can produce {mass_h2o:.2f} g of H₂O")

# Bonus: Calculate how much O₂ is needed
molar_mass_o2 = 32.00
moles_o2 = moles_h2 / 2  # 2:1 ratio H₂:O₂
mass_o2 = moles_o2 * molar_mass_o2
print(f"You need {mass_o2:.2f} g of O₂")
```

## Challenge Exercises Solutions

### Challenge 1: Ideal Gas Law
`P = (nRT)/V` with R = 0.0821 L·atm/(mol·K)

```python
# SOLUTION: Challenge 1
moles = float(input("Moles of gas (n): "))
temperature_k = float(input("Temperature (K): "))
volume_l = float(input("Volume (L): "))
R = 0.0821

pressure = (moles * R * temperature_k) / volume_l
print(f"Pressure = {pressure:.2f} atm")

# Convert to kPa for extra practice
pressure_kpa = pressure * 101.325
print(f"Pressure = {pressure_kpa:.1f} kPa")
```

### Challenge 2: Weight on Different Planets

```python
# SOLUTION: Challenge 2
mass_kg = float(input("Your mass (kg): "))

gravity_earth = 9.81
gravity_mars = 3.71
gravity_jupiter = 24.79

weight_earth = mass_kg * gravity_earth
weight_mars = mass_kg * gravity_mars
weight_jupiter = mass_kg * gravity_jupiter

print(f"\nYour weight on Earth:   {weight_earth:.1f} N")
print(f"Your weight on Mars:    {weight_mars:.1f} N")
print(f"Your weight on Jupiter: {weight_jupiter:.1f} N")

# Bonus: Comparison
print(f"\nYou would weigh {weight_mars/weight_earth*100:.0f}% of your Earth weight on Mars")
print(f"You would weigh {weight_jupiter/weight_earth:.1f} times your Earth weight on Jupiter")
```
