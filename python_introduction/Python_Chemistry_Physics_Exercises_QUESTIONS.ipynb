### Challenge 1: Ideal Gas Law
Calculate pressure: `PV = nRT` → `P = (nRT)/V`

Use R = 0.0821 L·atm/(mol·K)

```python
# Challenge 1
moles = float(input("Moles of gas (n): "))
temperature_k = float(input("Temperature (K): "))
volume_l = float(input("Volume (L): "))
R = 0.0821

# YOUR CODE HERE
pressure = (moles * R * temperature_k) / volume_l
print(f"Pressure = {pressure:.2f} atm")
```
